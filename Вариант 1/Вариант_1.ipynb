## Б)

```python
labels = {
    0: 'Нет 5G и NFC',
    1: 'Есть NFC, но нет 5G',
    2: 'Есть 5G, но нет NFC',
    3: 'Есть 5G и NFC'
}

conditions = [
    (df['has_5g'] == 0) & (df['has_nfc'] == 0),
    (df['has_5g'] == 0) & (df['has_nfc'] == 1),
    (df['has_5g'] == 1) & (df['has_nfc'] == 0),
    (df['has_5g'] == 1) & (df['has_nfc'] == 1)
]

df['category'] = np.select(conditions, labels.values(), default='Неизвестно')

summary = df.groupby('category', as_index=False)['battery_capacity'].mean()

fig_2 = px.bar(
    summary,
    x='category',
    y='battery_capacity',
    title='Влияние наличия 5G и NFC на среднюю ёмкость аккумулятора',
    labels={
        'category': 'Комбинация признаков',
        'battery_capacity': 'Средняя ёмкость, мАч'
    }
)

fig_2.show()
```
