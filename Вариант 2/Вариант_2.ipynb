# Задание 14

## А)

```python
fig_1 = px.histogram(
    df,
    x='bmi',
    y='expenses',
    histfunc='avg',
    title='Зависимость между индексом массы тела и медицинскими рассходами',
    labels={
        'bmi': 'Индекс массы тела (bmi)',
        'expenses': 'Медицинские расходы'
    },
    text_auto='.2s')

fig_1.update_layout(
    yaxis_title="Среднее медицинских расходов",
    xaxis_title="Индекс массы тела"
)

fig_1.show()
```

## Б)

```python
fig_2 = px.histogram(
    df,
    x='region',
    y='premium',
    title='Распределение страховой премии в зависиомсти от региона',
    text_auto='.2s')

fig_2.update_layout(
    yaxis_title="Сумма страховых премий",
    xaxis_title="Регионы"
)

fig_2.show()
```
